# 1. Importing Libraries and Datasets

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/amazon_ecommerce_1M.csv")
df.head()
```

# 2. Data Understanding and Data Cleaning

```python
df.describe()
```

```python
# Check missing values
df.isna().sum()
```

```python
# Check duplicated data
df.duplicated().sum()
```

```python
# Standardize number writings
df['price'] = df['price'].round(2)
df['discount'] = df['discount'].round(2)
df['final_price'] = df['final_price'].round(2)
df['rating'] = df['rating'].round(1)
df['seller_rating'] = df['seller_rating'].round(1)
df.head()
```

```python
# Fix inconsistencies in final prices
df['expected_final'] = df['price'] * (1 - df['discount'] / 100)
df['price_mismatch'] = (df['final_price'] - df['expected_final']).abs() > 1
# Decision: recalculate final_price or flag these rows
df['final_price'] = df['expected_final'].round(2)
df.head()
```

```python
# Count item with zero in stock
zero_stock = df[df['stock'] == 0]
len(zero_stock)
```

```python
df['purchase_date'] = pd.to_datetime(df['purchase_date'])
```

```python
df['location'] = df['location'].str.strip().str.title()
df.head()
```

```python
df.to_csv('amazon_ecommerce_cleaned.csv', index=False)
print("File 'amazon_ecommerce_cleaned.csv' has been created successfully.")
```
